### Data Visualization

#### City with highest attacks 

```python
data['City'].value_counts().to_frame().drop('Unknown').sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(15,8),color='#ffb500')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.title("Top 10 affected cities",fontsize=20)
plt.show()
```

#### Country with highest attacks

```python
plt.subplots(figsize=(15,8))
sns.barplot(x=data['Country'].value_counts()[:10].index,y=data['Country'].value_counts()[:10].values,palette='BuGn')
plt.title('Top 10 affected countries', fontsize = 20)
plt.xlabel('Country', fontsize = 15)
plt.ylabel('Number of attacks', fontsize = 15)
plt.xticks(rotation = 50)
plt.show()
```

### Region with most attacks

```python
data['Region'].value_counts().to_frame().sort_values('Region',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(15,8),color='#00a6a7')
plt.xticks(rotation = 50)
plt.xlabel("Region",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.title("Top 10 affected regions",fontsize=20)
plt.show()
```

#### Group with most attacks

```python
data[['Group','Killed']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('Killed',ascending=False).head(10).plot(kind='bar',color='#ffd500',figsize=(15,8))
plt.title("Top 10 Attacking groups",fontsize=20)
plt.xlabel("Terrorist group",fontsize=20)
plt.ylabel("Number of attacks",fontsize=20)
plt.show()
```

### Target types with most attacks

```python
plt.subplots(figsize=(15,8))
sns.countplot(x=data["Target_type"],order=data['Target_type'].value_counts().index,palette="viridis");
plt.xticks(rotation=90)
plt.xlabel("Type of Target",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.title("Most attacking types",fontsize=20)
plt.show()
```

### Attack Types

```python
data['AttackType'].value_counts().plot(kind='bar',figsize=(15,8),color='#00a6a7')
plt.xticks(rotation = 50)
plt.xlabel("Attack Type",fontsize=15)
plt.ylabel("Number of attacks",fontsize=15)
plt.title("Attack types over number of attacks",fontsize=20)
plt.show()
```

```python
data['Month'].value_counts()
plt.figure(figsize = (10, 6))
sns.countplot(x='Month', data = data)
plt.xticks(rotation=90)
plt.show()
```
